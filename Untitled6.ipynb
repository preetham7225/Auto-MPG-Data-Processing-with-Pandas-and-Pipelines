```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer
```

```python
# Step 1: Load the data
df = pd.read_csv('auto-mpg.csv')

# Display the first few rows of the dataframe
print("First few rows of the dataframe:")
print(df.head())

# Display the data types of each column
print("\nData types of each column before conversion:")
print(df.dtypes)

# Check for missing values
print("\nMissing values in each column before conversion:")
print(df.isnull().sum())

# Convert 'horsepower' to numeric (it might contain non-numeric values which need to be handled)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Check again for missing values after conversion
print("\nMissing values in each column after conversion:")
print(df.isnull().sum())

# Fill missing values in 'horsepower' column with the mean of the column
df['horsepower'].fillna(df['horsepower'].mean(), inplace=True)

# Verify the changes
print("\nData types of each column after handling missing values:")
print(df.dtypes)

# Check for missing values again to ensure none are left
print("\nMissing values in each column after handling missing values:")
print(df.isnull().sum())
```

```python
# Step 2: Split the data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.25, random_state=42)

# Print the shapes of the training and testing sets
print(f'\nTraining set shape: {train_df.shape}')
print(f'Testing set shape: {test_df.shape}')
```

```python
# Step 3: Visualize necessary columns in the data
# Plotting histograms for numeric columns to identify distribution types
numeric_cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']
for col in numeric_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

# Identify column types based on visual inspection
numeric_symmetric = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
numeric_skewed = ['cylinders']  # Assuming based on the histogram, modify if needed
categorical_cols = ['origin', 'car name']
```

```python
# Step 4: Build the pipeline
# Imputation and Standardization for symmetric numeric columns
numeric_symmetric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Imputation, Log transformation, and Standardization for skewed numeric columns
numeric_skewed_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('log_transform', FunctionTransformer(np.log1p)),
    ('scaler', StandardScaler())
])

# One hot encoding for categorical columns
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine all pipelines into a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num_sym', numeric_symmetric_pipeline, numeric_symmetric),
        ('num_skew', numeric_skewed_pipeline, numeric_skewed),
        ('cat', categorical_pipeline, categorical_cols)
    ]
)
```

```python
# Step 5: Train the pipeline on the training data
preprocessor.fit(train_df)

# Transform the training data and testing data
train_processed = preprocessor.transform(train_df)
test_processed = preprocessor.transform(test_df)
```

```python
# Step 6: Print the shape of the processed training data and processed testing data
print(f'\nProcessed training set shape: {train_processed.shape}')
print(f'Processed testing set shape: {test_processed.shape}')
```
